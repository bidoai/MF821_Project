```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
```

```python
raw_spot = pd.read_excel("mf821+project+data.xlsx",sheet_name=0)
raw_future = pd.read_excel("mf821+project+data.xlsx",sheet_name=1)
```

```python
names = raw_spot.columns
```

```python
spot_price = [names[x] for x in range(len(names)) if x % 7 == 1]
spot_bid = [names[x] for x in range(len(names)) if x % 7 == 3]
spot_ask = [names[x] for x in range(len(names)) if x % 7 == 5]

price_date = [names[x] for x in range(len(names)) if x % 7 == 0]
bid_date = [names[x] for x in range(len(names)) if x % 7 == 2]
ask_date = [names[x] for x in range(len(names)) if x % 7 == 4]
```

```python
raw_bid_prices = raw_spot[spot_bid]
raw_ask_prices = raw_spot[spot_ask]
raw_prices = raw_spot[spot_price]
```

```python
bid_dates = raw_spot[bid_date]
ask_dates = raw_spot[ask_date]
price_dates = raw_spot[price_date]
```

```python
col_bid = raw_bid_prices.columns
col_bid = col_bid[::-1]

col_ask = raw_ask_prices.columns
col_ask = col_ask[::-1]

col_price = raw_prices.columns
col_price = col_price[::-1]
```

```python
col_bid_date = bid_dates.columns
col_bid_date = col_bid_date[::-1]

col_ask_date = ask_dates.columns
col_ask_date = col_ask_date[::-1]

col_price_date = price_dates.columns
col_price_date = col_price_date[::-1]
```

```python
all_bids = pd.concat([ raw_bid_prices[col_bid[i]].dropna() for i in range(len(col_bid))])#.isna().sum()#.dropna()
all_asks = pd.concat([ raw_ask_prices[col_ask[i]].dropna() for i in range(len(col_ask))])
all_prices = pd.concat([ raw_prices[col_price[i]].dropna() for i in range(len(col_price))])
```

```python
all_bid_dates = pd.concat([ bid_dates[col_bid_date[i]].dropna() for i in range(len(col_bid_date))])
all_ask_dates = pd.concat([ ask_dates[col_ask_date[i]].dropna() for i in range(len(col_ask_date))])
all_price_dates = pd.concat([ price_dates[col_price_date[i]].dropna() for i in range(len(col_price_date))])
```

```python
p = pd.to_datetime(all_bid_dates)
agg = pd.DataFrame([all_bids.values, all_asks.values, all_prices.values]).T
```

```python
agg.columns = ["Bid", "Ask", "Price"]
agg["Mid"] = (agg["Bid"] + agg["Ask"]) / 2
agg = agg.set_index(p)
#agg
```

```python
agg_spot = agg.resample('1T').first().fillna(method="bfill")
#agg_spot
```

```python
names = raw_future.columns

future_price = [names[x] for x in range(len(names)) if x % 7 == 1]
future_bid = [names[x] for x in range(len(names)) if x % 7 == 3]
future_ask = [names[x] for x in range(len(names)) if x % 7 == 5]

price_date = [names[x] for x in range(len(names)) if x % 7 == 0]
bid_date = [names[x] for x in range(len(names)) if x % 7 == 2]
ask_date = [names[x] for x in range(len(names)) if x % 7 == 4]

raw_bid_prices = raw_future[future_bid]
raw_ask_prices = raw_future[future_ask]
raw_prices = raw_future[future_price]

bid_dates = raw_future[bid_date]
ask_dates = raw_future[ask_date]
price_dates = raw_future[price_date]

col_bid = raw_bid_prices.columns
col_bid = col_bid[::-1]

col_ask = raw_ask_prices.columns
col_ask = col_ask[::-1]

col_price = raw_prices.columns
col_price = col_price[::-1]

col_bid_date = bid_dates.columns
col_bid_date = col_bid_date[::-1]

col_ask_date = ask_dates.columns
col_ask_date = col_ask_date[::-1]

col_price_date = price_dates.columns
col_price_date = col_price_date[::-1]

all_bids = pd.concat([ raw_bid_prices[col_bid[i]].dropna() for i in range(len(col_bid))])#.isna().sum()#.dropna()
all_asks = pd.concat([ raw_ask_prices[col_ask[i]].dropna() for i in range(len(col_ask))])
all_prices = pd.concat([ raw_prices[col_price[i]].dropna() for i in range(len(col_price))])

all_bid_dates = pd.concat([ bid_dates[col_bid_date[i]].dropna() for i in range(len(col_bid_date))])
all_ask_dates = pd.concat([ ask_dates[col_ask_date[i]].dropna() for i in range(len(col_ask_date))])
all_price_dates = pd.concat([ price_dates[col_price_date[i]].dropna() for i in range(len(col_price_date))])

p = pd.to_datetime(all_ask_dates)
agg = pd.DataFrame([all_bids.values, all_asks.values, all_prices.values]).T

agg.columns = ["Bid", "Ask", "Price"]
agg["Mid"] = (agg["Bid"] + agg["Ask"]) / 2
agg = agg.set_index(p)
agg_future = agg.resample('1T').first().fillna(method="bfill")
#agg_future
```

```python
final_df = pd.merge(agg_spot, agg_future, how="outer", left_index=True, right_index=True).dropna()
final_df.columns = ["Spot_bid", "Spot_ask", "Spot_price", "Spot_mid", "Future_bid", "Future_ask", "Future_price", "Future_mid"]
final_df
```

```python
final_df.index
```

```python
# Index a certain day to feed into VAR
sample = final_df[["Spot_mid", "Future_mid"]]['2017-11-24']
```

```python
model = VAR(sample)
res = model.fit(1)
res.summary()
```
